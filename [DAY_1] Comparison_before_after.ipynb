```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

%matplotlib inline
#%matplotlib

# define colors
GRAY1, GRAY2, GRAY3 = '#231F20', '#414040', '#555655'
GRAY4, GRAY5, GRAY6 = '#646369', '#76787B', '#828282'
GRAY7, GRAY8, GRAY9 = '#929497', '#A6A6A5', '#BFBEBE'
BLUE1, BLUE2, BLUE3, BLUE4 = '#174A7E', '#4A81BF', '#94B2D7', '#94AFC5'
RED1, RED2 = '#C3514E', '#E6BAB7'
GREEN1, GREEN2 = '#0C8040', '#9ABB59'
ORANGE1 = '#F79747'

# 플롯 폰트 패밀리를 Arial로 설정
plt.rcParams['font.family'] = 'NanumGothic'
# 수학 텍스트 굵게 및 이탤릭체 폰트 패밀리를 Arial로 설정
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.bf'] = 'NanumGothic:bold'
matplotlib.rcParams['mathtext.it'] = 'NanumGothic:italic'

# 설문 응답
X = ['지루함', '좋지 않음', '보통', '관심 있음', '흥미로움']
Y = [11, 5, 40, 25, 19] # 프로그램 이전 응답 분포
Z = [12, 6, 14, 30, 38] # 프로그램 이후 응답 분포

# 새로운 플롯 생성
plt.figure(figsize=(7.45, 4.9), # 가로, 세로 크기 (인치 단위)
           dpi=110)             # 그림 해상도

# subplot 레이아웃 조정
plt.subplots_adjust(left=0, right=0.735, top=0.665, bottom=0.17)

# matplotlib에는 서브카테고리 바 플롯에 대한 자동 지원이 없음
def subcategorybar(X,  # 카테고리 이름
                   vals,  # 서브카테고리 값
                   base_colors,  # 각 서브카테고리의 기본 색상
                   special_colors,  # 서브카테고리 강조 색상
                   special_bars,  # 강조된 바 인덱스
                   width=0.9):  # 전체 서브카테고리 바 폭
    n = len(vals)
    _X = np.arange(len(X))
    for i in range(n):
        # i번째 서브카테고리 바 그리기
        bars = plt.bar(_X - width / 2. + i / float(n) * width,
                       vals[i],
                       width=width / float(n),
                       align="edge",  # 바를 왼쪽 모서리로 정렬
                       color=base_colors[i])  # 서브카테고리에 기본 색상 사용

        # 각 바에 수치 데이터 라벨 추가
        for bi, b in enumerate(bars):
            plt.text(b.get_x() + 0.05, b.get_height() - 3.5,
                     str(vals[i][bi]) + '%',
                     fontsize=12,
                     color='white')
        # 바 강조
        if special_bars[i] != None:
            for special_bar_i in special_bars[i]:
                bars[special_bar_i].set_color(special_colors[i])
    # x축에 카테고리 틱 위치 및 레이블 설정
    plt.xticks(_X, X)


# 서브카테고리 수직 바 플롯 그리기
subcategorybar(X,  # 응답
               [Y, Z],  # 프로그램 이전 및 이후 응답 분포
               [GRAY8, BLUE3],  # 프로그램 이전 및 이후 바 색상
               [GRAY7, BLUE2],  # 프로그램 이전 및 이후 강조 색상
               [[2], [3, 4]])  # 3번째 프로그램 이전 및 4번째, 5번째 프로그램 이후 바 강조

# 차트 테두리 제거
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# 눈금, 눈금 라벨 및 그리드 라인 모양 변경
plt.tick_params(bottom='off', left='off', labelleft='off')

# 응답 라벨 모양 설정
for item in plt.gca().get_xticklabels():
    item.set_fontsize(14)
    item.set_color(GRAY3)

# 플롯에 제목 추가
plt.text(-0.65, 61,
         '시범 프로그램은 성공적이었습니다' + ' ' * 27,
         fontsize=26,
         color='white',
         # 텍스트 주변에 사각형 상자 추가하여 배경색 설정
         bbox={'facecolor': GRAY7,  # 배경색
               'pad': 10,  # 패딩 추가
               'edgecolor': 'none'})  # 테두리 없음

# 설문 질문 추가
plt.text(-0.65, 52, '과학에 대해 어떻게 생각하십니까?', fontsize=17, 
         color=GRAY1)

# 데이터 소스를 설명하는 각주 추가
plt.text(-0.65, -12.5,
         '시범 프로그램 이전과 이후에 대해 100명의 학생을 대상으로 '
         '실시한 설문 조사를 기반으로 함 '
         '(두 설문 모두 100% 응답률).',
         fontsize=10.5,
         color=GRAY3)

# 플롯에 이전 노트 추가
plt.text(-0.52, 42,
         '$\\bf{프로그램 이전}$, 대다수의 아이들이\n'  # 굵은 텍스트를 위해 \\bf 사용
         '과학에 대해 그저 $\\mathit{보통}$으로 느꼈습니다.',  # 이탤릭체 텍스트를 위해 \\mathit 사용
         fontsize=14,
         linespacing=1.3,  # 글자 크기의 배수로 줄 간격 설정
         color=GRAY3)

# 플롯에 이후 노트 추가
plt.text(4.6, 10,
         '$\\bf{프로그램 이후}$\n'  # 굵은 텍스트를 위해 \\b 사용
         '아이들이\n'
         '더 $\\it{관심이 있고}$\n'  # 이탤릭체 텍스트를 위해 \\it 사용
         '$\\it{흥미로워졌습니다}$.\n',
         fontsize=14,
         linespacing=1.45,  # 글자 크기의 배수로 줄 간격 설정
         color=BLUE2)
```
