```python
import matplotlib
from matplotlib import pyplot as plt

%matplotlib inline
#%matplotlib
```

```python
# define colors
GRAY1, GRAY2, GRAY3 = '#231F20', '#414040', '#555655'
GRAY4, GRAY5, GRAY6 = '#646369', '#76787B', '#828282'
GRAY7, GRAY8, GRAY9 = '#929497', '#A6A6A5', '#BFBEBE'
BLUE1, BLUE2, BLUE3, BLUE4 = '#174A7E', '#4A81BF', '#94B2D7', '#94AFC5'
RED1, RED2 = '#C3514E', '#E6BAB7'
GREEN1, GREEN2 = '#0C8040', '#9ABB59'
ORANGE1 = '#F79747'
```

```python
# 플롯 폰트 패밀리를 Arial로 설정
plt.rcParams['font.family'] = 'NanumGothic'
# 수학 텍스트 굵게 및 이탤릭체 폰트 패밀리를 Arial로 설정
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.bf'] = 'NanumGothic:bold'
matplotlib.rcParams['mathtext.it'] = 'NanumGothic:italic'
```

```python
# 새로운 그림 생성
plt.figure(figsize=(7.45, 4.9), dpi=110)

# 차트 테두리 제거
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# 그림 제목 추가
plt.text(-0.15, 1.03,
         '시범 프로그램은 성공적이었습니다'+' '*27,
         fontsize=26,
         color='white',
         bbox={'facecolor': GRAY7, 'pad': 10, 'edgecolor': 'none'})

# 데이터 출처 각주
plt.text(-0.15, 0.04,
         '시범 프로그램 전후에 100명의 학생을 대상으로 한 설문 조사 '
         '(두 번의 조사 모두 100% 응답률).',
         fontsize=10.5,
         color=GRAY3)

# 그림에 설명 추가
plt.text(-0.15, 0.81,
         '시범 프로그램 이후,',
         fontsize=19,
         color=BLUE2)

# 그림에 설명 추가
plt.text(-0.15, 0.41,
         '$\\bf{68\%}$',
         fontsize=122,
         color=BLUE2)

# 그림에 설명 추가
plt.text(-0.15, 0.25,
         '$\\bf{아이들이 \ 과학에\ 대한 \ 관심을\ 표현했습니다,}$',
         fontsize=19,
         color=BLUE2)

# 그림에 설명 추가
plt.text(-0.15, 0.17,
         '프로그램에 참여하기 전 44%와 비교해보세요.',
         fontsize=19,
         color=GRAY7)

# 축 눈금과 라벨 제거
plt.xticks([])
plt.yticks([])

# 그림 표시
plt.show()
```
